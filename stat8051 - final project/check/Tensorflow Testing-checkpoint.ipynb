```python
# Importing our packages
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
```

```python
# Importing all our data
df = pd.read_csv('InsNova_train.csv')
df['pure_premium'] = df['claim_cost'] / df['exposure']
df['avg_cost'] = df['claim_cost'] / np.fmax(df['claim_count'], 1)
df['frequency'] = df['claim_count'] / df['exposure']
response_cols = ['exposure', 'claim_ind', 'claim_count', 'claim_cost', 'pure_premium', 'avg_cost', 'frequency']
df = pd.get_dummies(df, columns=['veh_body', 'gender', 'area'])
for i in ['veh_value', 'veh_age', 'dr_age']:
    df[i] = df[i] / df[i].max()
X, y = df.drop(response_cols, axis=1), df[response_cols]
X = X.drop('id', axis=1)
X['intercept'] = 1.0
#X, y = X.to_numpy(), y.to_numpy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, stratify=df['claim_count'])
```

```python
# Defining our Tweedie loss
def tweedie_loss(y_true, y_pred):
    p = 1.5
    loss = 2.0 * (tf.pow(y_true, 2-p) / ((1-p) * (2-p)) - y_true * tf.pow(y_pred, 1-p) / (1-p) + tf.pow(y_pred, 2-p) / (2-p))
    return tf.reduce_mean(loss, axis=-1)
```

```python
# Trying simple GLM
def create_model():
    model = tf.keras.Sequential([
        tf.keras.Input((25,)),
        tf.keras.layers.Dense(1, activation=tf.keras.activations.exponential)
    ])
    model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.005), loss=tweedie_loss)
    return model
tf_model = tf.keras.wrappers.scikit_learn.KerasRegressor(create_model, epochs=30, batch_size=64)
model.fit(X_train, y_train, epochs=20, validation_data=(X_test, y_test))
```

```python
# Creating our tf.keras network
model = tf.keras.Sequential([
    tf.keras.Input((24,)),
    tf.keras.layers.Dense(32),
    tf.keras.layers.LeakyReLU(0.1),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(32),
    tf.keras.layers.LeakyReLU(0.1),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(32),
    tf.keras.layers.LeakyReLU(0.1),
    tf.keras.layers.Dropout(0.3),
    tf.keras.layers.Dense(1, activation=tf.keras.activations.exponential)
])
model.summary()
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.001), loss=tweedie_loss)
model.fit(X_train, y_train, epochs=30, validation_data=(X_test, y_test))
```
